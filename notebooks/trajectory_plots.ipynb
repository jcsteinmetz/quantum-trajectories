```python
import sys
import os
sys.path.append(os.path.abspath('..'))

import numpy as np
from matplotlib import pyplot as plt
from sqt.qubit import Qubit
from sqt.controls import Rabi, Measurement
```

```python
# Create controls
rabi = Rabi(direction = [1,0,0], angle = np.pi, time_start = 0, time_end = 1000)
measurement = Measurement(direction = [0,0,1], tau = 100)

qubit = Qubit([0,1,0])
qubit.set_controls([rabi, measurement])

fig = qubit.plot_trajectories(n_trajectories = 10, time_start = 0, time_end = 1000, dt = 1, stochastic = True)
```
