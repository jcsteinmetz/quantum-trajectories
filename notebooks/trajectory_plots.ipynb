```python
import sys
import os
sys.path.append(os.path.abspath('..'))

import numpy as np
from sqt.qubit import Qubit
from sqt.controls import RabiDrive
from sqt.measurements import WeakMeasurement
```

```python
# Create controls
rabi = RabiDrive(direction = [1,0,0], angle = np.pi, time_start = 0, time_end = 1000, pulse_type = "sine squared")
m_z = WeakMeasurement(direction = [0,0,1], tau = 100, efficiency = 1)

qubit = Qubit([0,1,0])
qubit.set_controls([rabi])
qubit.set_measurements([m_z])

fig = qubit.plot_trajectories(n_trajectories = 10, time_start = 0, time_end = 1000, dt = 1, stochastic = True)
```
