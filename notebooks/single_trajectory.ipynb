```python
import sys
import os
sys.path.append(os.path.abspath('..'))

import numpy as np
from matplotlib import pyplot as plt
from sqt.qubit import Qubit
from sqt.drive import Rabi, Measurement
```

```python
rabi_drive = Rabi(rotation_axis = [1,0,0], rotation_angle = np.pi, time_start = 0, time_end = 1000)
measurement_drive = Measurement(measurement_axis = [0,0,1], tau = 100)

qubit = Qubit([0,1,0])
qubit.set_drives([rabi_drive, measurement_drive])

time, ens_avg_data = qubit.trajectory(time_start = 0, time_end = 1000, dt = 1, stochastic = False)
_, traj_data = qubit.trajectory(time_start = 0, time_end = 1000, dt = 1)
```

```python
plt.plot(time, ens_avg_data[0], 'r', label='x')
plt.plot(time, ens_avg_data[1], 'g', label='y')
plt.plot(time, ens_avg_data[2], 'b', label='z')

plt.plot(time, traj_data[0], 'r--', label='x')
plt.plot(time, traj_data[1], 'g--', label='y')
plt.plot(time, traj_data[2], 'b--', label='z')
plt.legend()
plt.ylim([-1,1])
```
