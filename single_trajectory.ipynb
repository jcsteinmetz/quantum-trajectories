```python
import numpy as np
from matplotlib import pyplot as plt
from qubit import Qubit
from drive import Rabi, Measurement
```

```python
rabi_drive = Rabi(rotation_axis = [1,0,0], rotation_angle = np.pi, time_start = 0, time_end = 200)
measurement_drive = Measurement(measurement_axis = [0,0,1], tau = 100)
measurement_drive_x = Measurement(measurement_axis = [1,0,0], tau = 100)

qubit = Qubit([1,0,0])
qubit.set_drives([measurement_drive])

time, traj_data = qubit.trajectory(time_start = 0, time_end = 1000, dt = 1, stochastic = False)
```

```python
plt.plot(time, traj_data[0], label='x')
plt.plot(time, traj_data[1], label='y')
plt.plot(time, traj_data[2], label='z')
plt.legend()
plt.ylim([-1,1])
```
